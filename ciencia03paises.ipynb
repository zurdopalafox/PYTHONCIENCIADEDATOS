```python
import pandas as pd
```

Creamos un Diccionario con Paises

```python
dataPaises = {
    'pais': ['España', 'Francia', 'Alemania', 'Italia', 'Portugal'],
    'capital': ['Madrid', 'París', 'Berlín', 'Roma', 'Lisboa'],
    'superficie': [505992, 643801, 357022, 301340, 92090],
    'idioma': ['español', 'francés', 'alemán', 'italiano', 'portugués']
}
dataPaises
```

Creamos un Data Frame con el diccionario

```python
dfpaises = pd.DataFrame(dataPaises)

dfpaises
```

Por ejemplo podemos dibujar solo una columna

```python
dfpaises[['capital']]
```

Podemos dibujar + de una

```python
dfpaises[['capital', 'idioma']]
```

Se pueden seleccionar las filas por su posicion(indice)
Primero incluido el segundo no.

```python
dfpaises[2:5]
```

Podemos sustituir el indice que viene por fefecto si quisieramos.

```python
dfpaises.index = ['FR', 'AL', 'IT', 'PT', 'ES']
dfpaises
```

podemos volver a crear nuevamente el indice por defecto

```python
dfpaises = dfpaises.reset_index()
dfpaises
```
