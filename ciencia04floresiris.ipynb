Vamos aa averiguar que tendriamos que hacer si tuvieramos valores nulos en nuestros datos.
Para trabajar con estos valores se utiliza numpy.

```python
import numpy as np
```

Incluimos valores nulos en las dos primeras filas: np.nan

```python
df.loc[:2,'petal_length'] = np.nan
df.head()
```

Los datos nulos son elementos que no se han rellenado en un dataframe.
Para tratarlos tenemos un par de opciones:
1) dropna() ---> Elimina la columna o fila con nulos
   Podemos indicar axis = 0 para filas
                   axis = 1 para las columnas 



```python
df_limpio = df.dropna(axis=0)
df_limpio.head()
```

```python
df.loc[:2,'petal_length'] = np.nan
df.head()
```

Podemos operar con los datos del DataFrame. Un null es igual que en BBDD no se tiene en cuenta.
En lugar de limpiar el Dataframe de nulos podemos modificar los valores nulos tambien.
Quitar el nulo y poner un valor por defecto. 
Para ello se utiliza: fillna(valor)

```python
df['petal_length'] = df['petal_length'].fillna(999)
df
```

Podemos crear agrupaciones para averiguar caracteristicas respecto a una columna. Esto es lo mismo que un Groupby en SQLServer.

```python
df_grupo = df.groupby('species')['width'].mean()
df_grupo
```
