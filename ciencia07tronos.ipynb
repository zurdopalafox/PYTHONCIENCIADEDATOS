```python
df = df.fillna({"allegiance": "unknown", "killer": "unknown", "killers_house": "unknown"})
df.info()
```

Podemos hacer de todo, por ejemplo quedarnos con datos de la temporada 1.

```python
df_temp1 =df[df["season"]== 1]
muertes_temp1 = df_temp1.shape[0]
print(f"Muertes en la temporada 1: {muertes_temp1}")
```

```python
muertes_totales = df.shape[0]
print(f"Muertes totales: {muertes_totales}")
```

```python
impacto_temp1 = round(muertes_temp1/muertes_totales * 100,2)
print(f"Impacto en la temporada 1: {impacto_temp1}%")
```

```python
df_temp1 =df[df["season"]== 1]
muertes_temp1 = df_temp1.shape[0]
print(f"Muertes en la temporada 1: {muertes_temp1}")
muertes_totales = df.shape[0]
print(f"Muertes totales: {muertes_totales}")
impacto_temp1 = round(muertes_temp1/muertes_totales * 100,2)
print(f"Impacto en la temporada 1: {impacto_temp1}%")
df_temp2 =df[df["season"]== 2]
muertes_temp2 = df_temp2.shape[0]
print(f"Muertes en la temporada 2: {muertes_temp2}")
muertes_totales = df.shape[0]
print(f"Muertes totales: {muertes_totales}")
impacto_temp2 = round(muertes_temp2/muertes_totales * 100,2)
print(f"Impacto en la temporada 2: {impacto_temp2}%")
df_temp3 =df[df["season"]== 3]
muertes_temp3 = df_temp3.shape[0]
print(f"Muertes en la temporada 3: {muertes_temp3}")
muertes_totales = df.shape[0]
print(f"Muertes totales: {muertes_totales}")
impacto_temp3 = round(muertes_temp3/muertes_totales * 100,2)
print(f"Impacto en la temporada 3: {impacto_temp3}%")
```

```python
data = {
    'Temporadas': ['Temporada1', 'Temporada2', 'Temporada3'],
    'Impacto (%)': [impacto_temp1, impacto_temp2, impacto_temp3]
}
df_impacto = pd.DataFrame.from_dict(data)
df_impacto.plot(x='Temporadas', y='Impacto (%)', kind='bar', legend=False)
plt.title('Impacto de las muertes por temporada')
plt.xlabel('Temporadas')
plt.ylabel('Impacto (%)')
plt.xticks(rotation=0)
plt.ylim(0, max(df_impacto['Impacto (%)']) + 10)
plt.show()
```
